# **Q-Q Plot**

```python
from statsmodels.graphics.gofplots import qqplot
import matplotlib.pylab as plt
from matplotlib import pyplot

data_numeric = data[
    ['Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)',
       'Metascore',"Success"]
]

# To check the likely distributions of numeric variables, we will use the QQ plot, descrptive analysis and histogram to determine the distribution
#ref https://www.statology.org/q-q-plot-python/
for c in data_numeric.columns[:]:
    data_numeric[c] = pd.to_numeric(data_numeric[c], errors="coerce")
    plt.figure(figsize=(6, 4))
    fig = qqplot(pd.to_numeric(data_numeric[c]), line="45")
    plt.xticks(fontsize=6)
    plt.yticks(fontsize=6)
    plt.xlabel("Theoretical quantiles", fontsize=10)
    plt.ylabel("Sample quantiles", fontsize=10)
    plt.title("Q-Q plot of {}".format(c), fontsize=15)
    plt.grid(True)
    plt.show()
```

# Observations:
Based on the Q-Q plots, we can infer the distributional characteristics for each variable:

1. *Runtime (Minutes)*:
   - The points follow the line closely at the center, suggesting that the middle values of the runtime are close to normally distributed.
   - The lower tail shows some deviation, indicating that the shortest movies are less than would be expected in a normal distribution.
   - The upper tail also shows deviation, with some outliers, indicating that the longest movies are longer than would be expected in a normal distribution.

2. *Rating*:
   - The lower tail deviates from the line, which suggests that there are more low-rated movies than what would be expected in a normal distribution.
   - The upper tail shows that high ratings also deviate from what a normal distribution would predict, possibly because there's a limit to the rating scale that causes a ceiling effect.

3. *Votes*:
   - The heavy concentration of points at the lower end and the long tail going off the line to the right indicates a right-skewed distribution.
   - This suggests that most movies have a small number of votes, but a few movies have a very large number of votes.

4. *Revenue (Millions)*:
   - Similar to the Votes plot, the Q-Q plot of Revenue is also highly right-skewed.
   - This indicates that most movies earn relatively little revenue, with a small number earning very high revenues.

5. *Metascore*:
   - The plot shows a deviation at both tails, which could mean the distribution is not normally distributed and could be bi-modal or have a different skewness or kurtosis than the normal distribution.

6. *Success*:
   - The data points in the Q-Q plot of Success are very tightly clustered around the line, except for a few outliers at the high end.
   - This could indicate that the 'Success' metric might have been transformed (e.g., standardized) and has a distribution close to normal except for these few extreme values.

In summary, the Q-Q plots suggest that 'Votes' and 'Revenue' are right-skewed, 'Runtime' and 'Rating' have some normal-like characteristics in the middle but deviate at the tails, 'Metascore' may not be normally distributed, and 'Success' is close to normal except for a few outliers. # **Removing outliers**

```python
import pandas as pd


Q1 = data['Votes'].quantile(0.25)
Q3 = data['Votes'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for what is considered an outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers

data_filtered= data[(data['Votes'] >= lower_bound) & (data['Votes'] <= upper_bound)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
# Using logistic regrssion algoritm to check the model accuracy after removing outlier, because from documentation, it has the most effect in accuracy after removing outliers

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve

x=data_filtered[data.columns[5:30]]
y=data_filtered["Success"]
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

logisticregression = LogisticRegression()
logisticregression.fit(x_train, y_train)
y_pred_lr_test = logisticregression.predict(x_test)

y_pred_lr_train = logisticregression.predict(x_train)

testacc_lr = accuracy_score(y_pred_lr_test, y_test)

trainacc_lr = accuracy_score(y_pred_lr_train, y_train)

print("Training Accuracy of Decision Tree: ", trainacc_lr)

print("Testing Accuracy in Logistic Regression: ", testacc_lr)
```

# Removing data randomly

```python
### Step 1: Remove Data Randomly
import numpy as np
import pandas as pd


df = data_numeric

# Function to remove data randomly
def remove_data_randomly(data, percentage):
    df_removed = data.copy()
    for column in df_removed.columns:
        # Calculate the number of data points to remove
        num_removal = int(np.floor(percentage * len(df_removed)))
        # Randomly choose indices
        indices = np.random.choice(df_removed.index, num_removal, replace=False)
        # Set chosen values to NaN
        df_removed.loc[indices, column] = np.nan
    return df_removed

# Remove 1%, 5%, and 10% of the data
df_1_percent = remove_data_randomly(df, 0.01)
df_5_percent = remove_data_randomly(df, 0.05)
df_10_percent = remove_data_randomly(df, 0.10)
```

```python
df_1_percent.isnull().sum()
```

Impute values back

```python
from sklearn.impute import SimpleImputer, KNNImputer

# Mean imputation
mean_imputer = SimpleImputer(strategy='mean')
df_mean_imputed_1 = mean_imputer.fit_transform(df_1_percent)

# Median imputation
median_imputer = SimpleImputer(strategy='median')
df_median_imputed_1 = median_imputer.fit_transform(df_1_percent)

# KNN imputation
knn_imputer = KNNImputer(n_neighbors=5) #taking 5 neighbors for now
df_knn_imputed_1 = knn_imputer.fit_transform(df_1_percent)

df_mean_imputed_5 = mean_imputer.fit_transform(df_5_percent)
df_median_imputed_5 = median_imputer.fit_transform(df_5_percent)
df_knn_imputed_5 = knn_imputer.fit_transform(df_5_percent)

df_mean_imputed_10 = mean_imputer.fit_transform(df_10_percent)
df_median_imputed_10 = median_imputer.fit_transform(df_10_percent)
df_knn_imputed_10 = knn_imputer.fit_transform(df_10_percent)
```

Find the imputation accuracy

```python
# Function to calculate the percentage error, bias, and variance
# Used chatgpt for the imputation methods and code
def evaluate_imputation(original, imputed):
    # Reshape if necessary
    if len(original.shape) == 1:
        original = original.reshape(-1, 1)
        imputed = imputed.reshape(-1, 1)

    residuals = original - imputed
    percentage_error = np.abs(residuals) / original
    bias = np.mean(residuals)
    variance = np.var(residuals)

    # Replace inf and NaN before taking mean for percentage error
    percentage_error = percentage_error[np.isfinite(percentage_error)].mean() * 100

    return percentage_error, bias, variance

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_mean_imputed_1[:, 0])
print(f"Mean Imputation 1%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_median_imputed_1[:, 0])
print(f"Median Imputation 1%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_knn_imputed_1[:, 0])
print(f"KNN Imputation 1%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_mean_imputed_5[:, 0])
print(f"Mean Imputation 5%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_median_imputed_5[:, 0])
print(f"Median Imputation 5%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_knn_imputed_5[:, 0])
print(f"KNN Imputation 5%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_mean_imputed_10[:, 0])
print(f"Mean Imputation 10%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_median_imputed_10[:, 0])
print(f"Median Imputation 10%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")

percentage_error, bias, variance = evaluate_imputation(data_numeric['Metascore'].values, df_knn_imputed_10[:, 0])
print(f"KNN Imputation 10%: Error={percentage_error:.2f}%, Bias={bias:.2f}, Variance={variance:.2f}")
```
